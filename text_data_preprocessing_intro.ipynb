```python
# можем использовать и (или создавать) различные маси

import numpy as np

def create_circular_mask(h, w, center=None, radius=None):

    if center is None: 
        center = (int(w/2), int(h/2))
    if radius is None: 
        radius = min(center[0], center[1], w-center[0], h-center[1])

    Y, X = np.ogrid[:h, :w]
    dist_from_center = np.sqrt((X - center[0]) ** 2 + (Y-center[1]) **2)

    mask = dist_from_center <= radius
    mask = ~mask
    
    return mask

mask = create_circular_mask(2000, 4000)
mask = 255 * mask.astype(int)

wc = WordCloud(width=4000, height=2000, max_words=100, 
               background_color='white', mask=mask).generate_from_frequencies(freq_tokens)

plt.figure(figsize=(20, 10))
plt.imshow(wc)
plt.tight_layout(pad=0)
plt.axis("off")
plt.show()
```

```python
import natasha
from natasha import (Segmenter, MorphVocab, NewsEmbedding, 
                    NewsMorphTagger, NewsSyntaxParser, NewsNERTagger, PER, NamesExtractor, Doc)
```

Распознаёт три типа именнованных сущностей: __PER__ - персоны, __OGR__ - организации, __LOC__ - локации

```python
raw_text = ' '.join(sentences)
doc = Doc(raw_text)
```

Поиск именованных сущностей:

```python
segmenter = Segmenter() 
emb = NewsEmbedding()
ner_tagger = NewsNERTagger(emb)
```

```python
doc.segment(segmenter) 
doc.tag_ner(ner_tagger)
doc.ner.print() 
```

```python
class color:
    
    PURPLE = '\033[95m'
    CYAN = '\033[96m'
    DARKCYAN = '\033[36m'
    BLUE = '\033[94m'
    GREEN = '\033[92m'
    YELLOW = '\033[93m'
    RED = '\033[91m'
    BOLD = '\033[1m'
    UNDERLINE = '\033[4m'
    END = '\033[0m'

orgs = [span.text for span in doc.spans if span.type == 'ORG']
orgs = list(set(orgs))
print(color.BOLD + color.GREEN + 'Организации:' + color.END, ', '.join(orgs))
print()

locs = [span.text for span in doc.spans if span.type == 'LOC']
locs = list(set(locs))
print(color.BOLD + color.BLUE + 'Локации:' + color.END, ', '.join(locs))
print()

pers = [span.text for span in doc.spans if span.type == 'PER']
pers = list(set(pers))
print(color.BOLD + color.BLUE + 'Персоны:' + color.END, ', '.join(pers))
print()
```

Определение частей речи в тексте

```python
segmenter = Segmenter() #для разделения текста  
emb = NewsEmbedding() # получение emb от слова
morph_tagger = NewsMorphTagger(emb) # морфологический разбор  

doc.segment(segmenter) # разделяем текст
doc.tag_morph(morph_tagger)# морфологический разбор  
doc.sents[0].morph.print()# морфологический разбор  для 1 предложения 
```
